```python
import pandas as pd
from sklearn.preprocessing import StandardScaler

# Memuat data transaksi pelanggan
data = {
    'jumlah_pembelian': [5, 20, 15, 10, 7, 25, 30, 12, 10, 8],
    'total_belanja': [500, 2500, 1500, 1000, 700, 3000, 3500, 1200, 1000, 800],
    'frekuensi_pembelian': [1, 5, 4, 2, 1, 6, 7, 3, 2, 2]
}
df = pd.DataFrame(data)

# Normalisasi data
scaler = StandardScaler()
data_normalized = scaler.fit_transform(df)
```

```python
from sklearn.cluster import KMeans
import matplotlib.pyplot as plt

# Menentukan jumlah cluster (k)
kmeans = KMeans(n_clusters=3, random_state=0)

# Menerapkan algoritma K-Means
df['cluster'] = kmeans.fit_predict(data_normalized)
```

```python
# Visualisasi hasil clustering
plt.scatter(df['total_belanja'], df['frekuensi_pembelian'], c=df['cluster'], cmap='viridis')
plt.xlabel('Total Belanja')
plt.ylabel('Frekuensi Pembelian')
plt.title('Segmentasi Pelanggan')
plt.show()

# Menghitung Sum of Squared Errors (SSE)
sse = kmeans.inertia_
print(f'Sum of Squared Errors (SSE): {sse}')
```

```python
# Menampilkan karakteristik masing-masing cluster
for cluster in range(3):
    print(f'\nCluster {cluster}:')
    print(df[df['cluster'] == cluster].describe())
```
